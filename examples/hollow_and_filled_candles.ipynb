# HOLLOW and FILLED CANDLESTICKS


### To best understand Hollow and Filled Candlesticks, we compare them to Traditional Candlesticks:

---

# Traditional Candlestick Charts:
- Traditional candlestick charts **contain only TWO Candle Colors**.  These two colors represent whether the 
  - Close is GREATER than the Open (**up** color), or the
  - Close is LESS than the Open (**down** color).
- **Example** Color Combinations:

| <h2>`Black (Close > Open)`<br><br>`Red   (Close < Open)`</h2> | <h2>`White (Close > Open)`<br><br>`Black (Close < Open)`</h2> | <h2>`Green (Close > Open)`<br><br>`Red   (Close < Open)`</h2> |
| :---- | :---- | :---- |
|  |  |  |

---
# Hollow and Filled Candlestick Charts:

### Hollow and Filled Candlesticks differ from Traditional Candlesticks in two imporant ways:

**1.** The **TWO** COLORS ***now*** represent the ***Change (up/down) FROM the PREVIOUS CLOSE*** (*not* from the Open).

**2.** The  **Change from the Open** is instead represented by whether or not the candle body is 
- **HOLLOW**: Close **UP**&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; from the **Open**, or
- **FILLED**: &nbsp;&nbsp;&nbsp;Close **DOWN** from the **Open**

# A NOTE of WARNING:
## Some color combinations display "hollow and filled" candlesticks better than others.

## Traditional Candlestick Styles:
- Sometimes, in traditional candlesticks, one candle color is ***the same or very similar to the background (or face) color of the plot***.  
- In such a case, throughout the financial industry, it is common to ***make the edge color different from, and contrasting with, the candle body*** (so the candle will stand out against the background (face) of the plot).
- In mplfinance, you can see this in the following styles: 'classic', 'default', 'mike', 
and 'blueskies':

````python
df = pd.read_csv('data/yahoofinance-SPY-20200901-20210113.csv',index_col=0,parse_dates=True).iloc[5:40]
kwargs=dict(volume=True,tight_layout=True,figscale=0.5)
for s in ['classic','default','mike','blueskies']:
    mpf.plot(df,**kwargs,style=s,type='candle',title=s)
````

---

# A NOTE of WARNING continued:

## In Hollow and Filled Candlesticks,<br> ***the edge and wick color are significant***<br>(representing up or down),
and therefore *cannot* be used *as an outline* to distinguish the candle from the background.

Thus, for `type='hollow_and_filled'`, some solid candles may blend in or be difficult to distinguish from the background of the plot, while hollow candles may appear to be the wrong color:

````python
for s in ['classic','default','mike','blueskies']:
    mpf.plot(df,**kwargs,style=s,type='hollow_and_filled',title=s)
````

---

# A NOTE of WARNING continued:

## In Hollow and Filled Candlesticks,<br> ***the edge and wick color are significant***
and therefore cannot be used as an outline to distinguish the candle from the background.

## The simplest way to fix this problem, while using one of these styles, is to make the `facecolor` of the plot darker, or lighter, or a different color:

### Here is an example of darkening, or coloring, the `facecolor`:


````python
kwargs=dict(type='hollow_and_filled',volume=True,tight_layout=True,figscale=0.6)
face_color_styles = [
    'default',
    mpf.make_mpf_style(base_mpf_style='default',facecolor='#b9c7df',gridstyle='dotted'),
    mpf.make_mpf_style(base_mpf_style='default',facecolor='#95ced0',gridstyle=''),
    mpf.make_mpf_style(base_mpf_style='default',facecolor='#56b0b3',gridstyle='dotted')
]

for s in face_color_styles:
    mpf.plot(df,**kwargs,style=s)
````

---

# A NOTE of WARNING continued:

## And here is an example of lightening the `facecolor`:


````python
kwargs=dict(type='hollow_and_filled',volume=True,tight_layout=True,figscale=0.6)
face_color_styles = [
    'mike',
    mpf.make_mpf_style(base_mpf_style='mike',facecolor='#484848'),
    mpf.make_mpf_style(base_mpf_style='mike',facecolor='#606060'),
]

for s in face_color_styles:
    mpf.plot(df,**kwargs,style=s)
````
