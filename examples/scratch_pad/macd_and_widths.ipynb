#### Using this dataframe, we can of course plot a basic ohlc or candlestick plot:



```python
df = idf.loc['2011-08-15':'2011-11-30',:]
df = idf.loc['2011-07-01':'2011-12-30',:]
```

```python
import mplfinance as mpf

mpf.plot(df,volume=True)#,widths_adjust=False)
```

```python
df.shape
df.columns
```

```python
exp12 = df['Close'].ewm(span=12, adjust=False).mean()
exp26 = df['Close'].ewm(span=26, adjust=False).mean()
```

```python
macd = exp12 - exp26
```

```python
signal = macd.ewm(span=9, adjust=False).mean()
hist = macd - signal
```

```python
#mpf.available_styles()
```

```python
apds = [mpf.make_addplot(exp12,color='fuchsia'),
        mpf.make_addplot(exp26,color='blue'),
        mpf.make_addplot(hist,type='bar',width=0.6,panel=1,color='dimgray',alpha=1),
        mpf.make_addplot(macd,panel=1,color='lime'),#,secondary_y=True),
        mpf.make_addplot(signal,panel=1,color='c')#,secondary_y=True),
       ]
apds = [mpf.make_addplot(exp12,color='lime'),
        mpf.make_addplot(exp26,color='c'),
        mpf.make_addplot(hist,type='bar',width=0.7,panel=1,color='dimgray',alpha=1,secondary_y=False),
        mpf.make_addplot(macd,panel=1,color='fuchsia',secondary_y=True),
        mpf.make_addplot(signal,panel=1,color='b',secondary_y=True),
       ]

mpf.plot(df,type='candle',addplot=apds,figscale=1.1,figratio=(8,5),
         style='blueskies',volume=True,volume_panel=2,panel_ratios=(6,3,2))#,widths_adjust=False)
mpf.plot(df,type='candle',addplot=apds,figscale=1.1,figratio=(8,6.5),
         style='default')#,widths_adjust=False)#,panel_ratios=(7,5))
```

```python
s = mpf.make_mpf_style(base_mpf_style='default',rc={'figure.facecolor':'lightgray'})
s = 'classic'
s = mpf.make_mpf_style(base_mpf_style='classic',rc={'figure.facecolor':'lightgray'})
```

```python
apds = [mpf.make_addplot(exp12,color='lime',panel=1),
        mpf.make_addplot(exp26,color='c',panel=1),
        mpf.make_addplot(hist,type='bar',width=0.6,panel=0,color='dimgray',alpha=1,secondary_y=False),
        mpf.make_addplot(macd,panel=0,color='fuchsia',secondary_y=True),
        mpf.make_addplot(signal,panel=0,color='b',secondary_y=True),
       ]

s = mpf.make_mpf_style(base_mpf_style='classic',rc={'figure.facecolor':'lightgray'})

mpf.plot(df,type='candle',addplot=apds,figscale=1,figratio=(8,6.5),
         style=s,panel_ratios=(4,6,3),volume=True,volume_panel=2,main_panel=1
         ,tight_layout=True,title='MACD')

mpf.plot(df,type='candle',addplot=apds,figscale=1,figratio=(8,6.5),
         style=s,panel_ratios=(4,6,3),volume=True,volume_panel=2,main_panel=1
         ,tight_layout=True,title='MACD',scale_width_config=dict(volume=2.1,candle=3))

retw = {}
mpf.plot(df,type='candle',addplot=apds,figscale=1,figratio=(8,6.5),
         style=s,panel_ratios=(4,6,3),volume=True,volume_panel=2,main_panel=1,
         tight_layout=True,title='MACD',width_adjuster_version='v0',
         scale_width_config=dict(candle=5.5,volume=0.1),
         #update_width_config=dict(candle_width=1,volume_linewidth=0.1),
         update_width_config=dict(volume_linewidth=0.1),
         return_width_config=retw)

print('retw=',retw,'\n')

mpf.plot(df,type='candle',addplot=apds,figscale=1.2,figratio=(8,6.5),
         style=s,panel_ratios=(4,6,3),volume=True,volume_panel=2,main_panel=1,
         savefig=dict(fname='tight.png',bbox_inches='tight'),
         tight_layout=True,title='MACD')
```

```python
from PIL import Image

im = Image.open('tight.png')

import matplotlib.pyplot as plt

def plotimg(im):
    fig,ax = plt.subplots()
    ax.set_axis_off()
    ax.imshow(im)
    plt.show()

plotimg(im)
```
